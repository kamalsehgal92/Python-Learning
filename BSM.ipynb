```python
#BSM model to value European equity
import math
import pandas as pd
import numpy as np
from scipy.stats import norm

def option_pricing_model (spot, strike, rate, sigma, maturity, option_type):
    d1 = (math.log(spot / strike) + (rate + 0.5 * sigma ** 2) * maturity) / (sigma * math.sqrt(maturity))
    d2 = (math.log(spot / strike) + (rate - 0.5 * sigma ** 2) * maturity) / (sigma * math.sqrt(maturity))

    print(d1, d2)

    nd1 = norm.cdf(d1)
    nd2 = norm.cdf(d2)
    print(nd1, nd2)

    if option_type == 'call':
        price = spot * nd1 - strike * math.exp(-rate * maturity) * nd2

    elif option_type == 'put':
        price = strike * math.exp(-rate * maturity) * (1 - nd2) - spot * (1 - nd1)
    
    return price


spot = 200
strike = 100
rate = 0.05
sigma = 0.2
maturity = 1
option_type = 'call'

price = option_pricing_model(spot, strike, rate, sigma, maturity, option_type)

print(f"The price of the {option_type} option is: {price:}") 
```
